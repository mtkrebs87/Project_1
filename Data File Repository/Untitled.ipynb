```python

data = pd.read_csv('your_data.csv')

# Extract the borough names
boroughs = data['Borough']

# Extract and clean the median rent values by removing '$' and converting to float
years = [str(year) for year in range(2010, 2024)]
for year in years:
    data[year] = data[year].str.replace('$', '').astype(float)

# Sort the data by median rent in descending order for each year
sorted_data = data[years].apply(pd.to_numeric).sort_values(by=years, axis=1, ascending=False)

# Create a bar chart for median rental prices by borough
plt.figure(figsize=(12, 6))
bar_width = 0.15
index = range(len(boroughs))

for i, year in enumerate(years):
    median_rent = sorted_data[year]
    plt.bar([x + i * bar_width for x in index], median_rent, width=bar_width, label=year)

plt.xlabel('Borough')
plt.ylabel('Median Rent')
plt.title('Median Rental Prices by Borough (2010-2023, Descending Order)')
plt.xticks([x + bar_width * (len(years) - 1) / 2 for x in index], boroughs)
plt.legend(title='Year')
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```
