```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

df_inventory = pd.read_csv("../Resources/rentalInventory_All.csv")
df_inventory.head()
```

```python
# Group the data in the DataFrame by the "Borough" column
grouped = df_inventory.groupby("Borough")
# Calculate the sum of each borough
Borough_sums = grouped.sum()
# Reset the index of the new DataFrame, which replaces the original index with a default integer index
Borough_sums_df = Borough_sums.reset_index()
# Drop the "areaName" and "areaType" columns from the DataFrame, leaving only the borough names and their sums
Borough_sums_df = Borough_sums_df.drop(columns=["areaName", "areaType"])
Borough_sums_df.head()
```

```python
# Group the data in the 'Borough_sums_df' DataFrame by the year 
df_yearly_sum = Borough_sums_df.groupby(Borough_sums_df.columns.str.extract('(\d{4})', expand=False), axis=1).sum()
# Convert the column names to integers assuming they represent years
df_yearly_sum.columns = df_yearly_sum.columns.astype(int)
# Insert a new column "Borough" at the beginning of the DataFrame
df_yearly_sum.insert(0, "Borough", Borough_sums_df["Borough"])
df_yearly_sum.head()
```

```python
# Save the DataFrame as a CSV file in the "Data File Repository" directory one level up
output_folder = '../Data File Repository'
output_file_path = os.path.join(output_folder, 'Rental_Inventory_DF.csv')
os.makedirs(output_folder, exist_ok=True)  # Create the folder if it doesn't exist
df_yearly_sum.to_csv(output_file_path, index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'2010': [3087, 59006, 441207, 19596, 55]}
boroughs = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island']
df_yearly_sum = pd.DataFrame(data, index=boroughs)

# Extract the '2010' data for all 'Borough' levels
values = df_yearly_sum['2010']
values.index.name = 'Borough'  # Optional: Set the index name for the labels

# Create a custom color palette for the slices using named colors
colors = ['lightcoral', 'lightskyblue', 'lightgreen', 'lightpink', 'yellow']

# Reduce the explosion of the first slice (optional)
explode = (0.05, 0, 0, 0.1, 0)

# Create a pie chart with adjusted labels and figure size
plt.figure(figsize=(8, 6))  # Adjust the figure size
plt.pie(values, labels=None, autopct='%1.1f%%', pctdistance=0.75, colors=colors, startangle=90, explode=explode)

# Add a title (adjust for 2010)
plt.title('Distribution of Data in 2010 by Borough')

# Make the pie chart a perfect circle
plt.axis('equal')

# List labels on both sides with adjusted distance
plt.legend(values.index, loc='center left', bbox_to_anchor=(0.9, 0.5))

# Show the pie chart
plt.show()
```

```python
data = {'2023': [18069, 296376, 481479, 113475, 742]}
boroughs = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island']
df_yearly_sum = pd.DataFrame(data, index=boroughs)

values = df_yearly_sum['2023']
values.index.name = 'Borough'

colors = ['lightcoral', 'lightskyblue', 'lightgreen', 'lightpink', 'yellow']
explode = (0.05, 0, 0, 0, 0)

# Adjust the labeldistance to move the labels further from the center
labeldistance = 1.1

plt.figure(figsize=(8, 6))
plt.pie(values, labels=None, autopct='%1.1f%%', pctdistance=0.5, labeldistance=labeldistance, colors=colors, startangle=90, explode=explode)

plt.title('Distribution of Data in 2023 by Borough')
plt.axis('equal')

plt.legend(values.index, loc='center left', bbox_to_anchor=(0.9, 0.5))

plt.show()
```
