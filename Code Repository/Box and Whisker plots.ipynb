```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Read the data from the CSV file
file_path = '../Data File Repository/Combined_Rental_Median_Price_Inventory.csv'
df = pd.read_csv(file_path)
```

```python
# Extract the columns related to total rentals for all years
year_columns = [col for col in df.columns if col.endswith('Total Rentals')]
```

```python
# Create a box and whisker plot with a logarithmic scale on the y-axis
plt.figure(figsize=(12, 6))
plt.boxplot([df[col].dropna() for col in year_columns], vert=False, labels=year_columns)
plt.title('Box and Whisker Plot of Total Rentals by Year (Log Scale)')
plt.xlabel('Number of Rentals')
plt.ylabel('Year')
plt.xscale('log')  # Apply a logarithmic scale to the x-axis
```

```python
# Specify the path to save the PNG file
output_directory = '../Charts and Graphs Output'
output_filename = 'Total_yearly_inventory_box_and_whisker.png'
output_path = os.path.join(output_directory, output_filename)
```

```python
# Ensure the output directory exists
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
```

```python
# Save the figure directly using the Figure object
fig = plt.gcf()
fig.savefig(output_path, dpi=300)
```

```python
# Close the figure to release resources
plt.close(fig)
```

```python
# Show the plot
plt.show()
```
