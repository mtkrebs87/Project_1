```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Read the data from the CSV file
file_path = '../Data File Repository/Combined_Rental_Median_Price_Inventory.csv'
df = pd.read_csv(file_path)
```

```python
# Extract the columns related to total rentals for all years
year_columns = [col for col in df.columns if col.endswith('Total Rentals')]
```

```python
# Create a box and whisker plot with a logarithmic scale on the y-axis
plt.figure(figsize=(12, 6))
plt.boxplot([df[col].dropna() for col in year_columns], vert=False, labels=year_columns)
plt.title('Box and Whisker Plot of Total Rentals by Year (Log Scale)')
plt.xlabel('Number of Rentals (Scaled)')
plt.ylabel('Year')
plt.xscale('log')  # Apply a logarithmic scale to the x-axis

# Create the folder if it doesn't exist
output_folder = '../Charts and Graphs Output'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    
# Save the plot as a PNG file in the "Charts and Graphs Output" folder
output_file_path = os.path.join(output_folder, 'inventory_box_plot.png')
plt.savefig(output_file_path)
```

```python
# Show the plot
plt.show()
```

```python
# code to create a box plot for the year 2023 referencing median rent price
median_rent_columns = [col for col in df.columns if col.endswith('Median Rent')]
median_rent_values = []
```

```python
# Clean and convert median rent values to numerical format
for col in median_rent_columns:
    median_rent_values.append(df[col].apply(lambda x: float(str(x).replace('$', '').replace(',', '').replace('nan', '0'))))

plt.figure(figsize=(12, 6))
plt.boxplot(median_rent_values, vert=False, labels=median_rent_columns)
plt.title('Box and Whisker Plot of Median Rent by Year')
plt.xlabel('Median Rent in USD')
plt.ylabel('Year')

# Save the plot for 2023 as a PNG file
output_file_path_median_rent = os.path.join(output_folder, 'boxplot_median_rent_all_years.png')
plt.savefig(output_file_path_median_rent)
```
